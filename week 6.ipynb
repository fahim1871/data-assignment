At first we import necessary libraries and call the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('supermarket_sales.csv')
print('This is a sample of the dataset:')
print(df)
print('This is a description of the dataset:')
print(df.describe())
print('This dataset shows the sales information of a supermarket. It has the data of sales of different branches of a supermarket in different cities.')
print('The dataset has',df.shape[0],'rows ')
print('The dataset has',df.shape[1],'columns ')
print('Here is short description of the columns:')
print(df.info())
```

2. Basic Plots (40 Marks)
 - Use Matplotlib to create the following visualizations from your chosen dataset:
   -   Line Plot: Show a trend or progression of a variable over time or another index. (10 Marks)
   -   Bar Chart: Compare categories or groups. (10 Marks)
   -   Scatter Plot: Illustrate relationships between two numeric variables. (10 Marks)
   -   Box Plot: Summarize the distribution and identify outliers of a numeric variable. (10 Marks)


```python
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(15,8))
plt.plot(daily_sales['Date'], daily_sales['Total'], marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
print('This is a line plot:')
plt.show()
sales_by_product_line = df.groupby('Product line')['Total'].sum().sort_values()
plt.figure(figsize=(15,8))
plt.bar(sales_by_product_line.index, sales_by_product_line.values,)
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
print('this is a bar plot:')
plt.show()
plt.figure(figsize=(15, 6))
plt.scatter(df['Unit price'], df['Total'], alpha=.5)
plt.title('Unit Price vs Total Sales')
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
print('This is a scatter plot:')
plt.show()
plt.figure(figsize=(12, 10))
plt.boxplot([df['Total']])
plt.title('Total Sales')
plt.xlabel('Total Sales' )
plt.show()
```

3. Customizing Plots (20 Marks)
 - Add meaningful titles, labels, and legends to all the plots. (10 Marks)
 - Experiment with colors and markers to make your plots more readable and visually appealing. (10 Marks)


```python
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(15,8))
plt.plot(daily_sales['Date'], daily_sales['Total'], marker='.',linestyle='-',color='red',label='Total sales')
plt.title('Total Sales Over Time',fontsize='16')
plt.xlabel('Date',fontsize='14')
plt.ylabel('Total Sales',fontsize='14')
plt.legend()
print('This is a line plot:')
plt.show()
sales_by_product_line = df.groupby('Product line')['Total'].sum().sort_values()
plt.figure(figsize=(15,8))
plt.bar(sales_by_product_line.index, sales_by_product_line.values,color='lightcoral', edgecolor='black')
plt.title('Total Sales by Product Line',fontsize='16')
plt.xlabel('Total Sales',fontsize='14')
plt.ylabel('Product Line',fontsize='14')
print('this is a bar plot:')
plt.show()
plt.figure(figsize=(15, 6))
plt.scatter(df['Unit price'], df['Total'],marker='.', alpha=.5,color='blue', edgecolor='black', s=50, label='Sales Data Points')
plt.title('Unit Price vs Total Sales',fontsize='16')
plt.xlabel('Unit Price',fontsize='14')
plt.ylabel('Total Sales',fontsize='14')
plt.legend()
print('This is a scatter plot:')
plt.show()
plt.figure(figsize=(10,5))
plt.boxplot([df['Total']],patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'),  medianprops=dict(color='red'))
plt.title('Total Sales')
plt.xlabel('Total Sales' )
print('This is a box plot:')
plt.show()
```

4. Insights and Observations (15 Marks)
For each plot:

```python
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(15,8))
plt.plot(daily_sales['Date'], daily_sales['Total'], marker='.',linestyle='-',color='red',label='Total sales')
plt.title('Total Sales Over Time',fontsize='16')
plt.xlabel('Date',fontsize='14')
plt.ylabel('Total Sales',fontsize='14')
plt.legend()
print('This is a line plot:')
plt.show()
```

From the line plot it can be seen that there is no definite pattern on the total sales with time.But it seems that dales was the lowest around mid February and the highest around mid of March. At the beginning of the year the was a decline of sales but it gradually got up. It can be said that there were no constant sales over the time period as there are very frequent crest and troughs.

```python
sales_by_product_line = df.groupby('Product line')['Total'].sum().sort_values()
plt.figure(figsize=(15,8))
plt.bar(sales_by_product_line.index, sales_by_product_line.values,color='lightcoral', edgecolor='black')
plt.title('Total Sales by Product Line',fontsize='16')
plt.xlabel('Total Sales',fontsize='14')
plt.ylabel('Product Line',fontsize='14')
print('this is a bar plot:')
plt.show()
```

From the barplot is seems that except for health and beauty products all other products were sold about the same amount. Health and beauty products were sold the least whereas food and beverages sold the most.

```python
plt.figure(figsize=(15, 6))
plt.scatter(df['Unit price'], df['Total'],marker='.', alpha=.5,color='blue', edgecolor='black', s=50, label='Sales Data Points')
plt.title('Unit Price vs Total Sales',fontsize='16')
plt.xlabel('Unit Price',fontsize='14')
plt.ylabel('Total Sales',fontsize='14')
plt.legend()
print('This is a scatter plot:')
plt.show()
```

The scatterplot shows a linear trend between unit prices and total sales.Taht means thay have a strong linear relationship.The plot suggests that higher-priced products tend to generate higher total sales. Which may not always be correct.

```python
plt.figure(figsize=(10,5))
plt.boxplot([df['Total']],patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'),  medianprops=dict(color='red'))
plt.title('Total Sales')
plt.xlabel('Total Sales' )
print('This is a box plot:')
plt.show()
```

In the box plot the median is around 250.Thus the majority of sales fall between approximately 100 and 450. The 75% mark is around 450.And the 25% is around 100. It is seen that the concentration of sales is towards the lower end.
